```python
#Importing Package 
import sys
print('Python: {}'.format(sys.version))
import numpy
import pandas
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
import numpy as np
import warnings
```

```python
#Reading IRIS Dataset in Pandas Dataframe
url="datasets/Iris.csv"
iris_dataset = pandas.read_csv(url)
iris_dataset.head(5)
```

```python
#findout unique classification/type of iris flower.
label_types = iris_dataset['Species'].unique()
print(label_types)
```

```python
label = iris_dataset['Species']
print(label)
```

```python
features = list(zip(iris_dataset['SepalLengthCm'], iris_dataset['SepalWidthCm'], iris_dataset['PetalLengthCm'], iris_dataset['PetalWidthCm']))
print(features)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=3)

# Train the model using the training sets
model.fit(features,label)

#Predict Output
predicted= model.predict([ (5.1, 3.5, 1.4, 0.2) ]) # 
print(predicted)
```

```python
predicted= model.predict([ (5.9, 3.0, 5.1, 1.8) ]) #
print(predicted)
```

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.3,random_state=109) # 70% training and 30% test
```

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
